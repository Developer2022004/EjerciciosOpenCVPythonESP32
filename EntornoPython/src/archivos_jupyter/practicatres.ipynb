```python
import matplotlib.pyplot as plt
```

```python
x = [10,50,20,80]
y = [10,20,30,40]
#Inicializamos la grafica para cargar los datos
plt.plot(y,x)
plt.show()
```

```python
plt.plot(y,x)
#Para nombrar el eje x de la grafica
plt.xlabel('eje X')
#Para nombrar el eje y de la grafica
plt.ylabel('eje y')

#Para asignar un titulo a la grafica
plt.title('Ejemplo de grafico de lineas')
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [100,50,80,40]
plt.bar(x1,y1)
plt.xlabel('Letras')
plt.ylabel('Valores')
plt.title('Ejemplo de grafico de barras')
plt.show()
```

```python
x2 = ['A','B','C','D']
y2 = [100,50,80,40]

plt.pie(y2,labels=x2,autopct='%1.1f%%')
'''
autopct="
% inicia el formato
1 indica el ancho minimo del numero
.1 especifica la cantidad de decimales
f especifica que debe de ser de tipo flotante
%% escapa el caracter de %
'''
plt.title('Grafica de pie')
plt.axis('equal') #Busca garantizar que sea un grafico circular
plt.show()
```

```python
x3 = [100,50,80,40]
y3 = [10,50,20,40]
z3 = [18,50,20,100]

figura = plt.figure()
grafico = figura.add_subplot(111,projection='3d')
grafico.plot(x3,y3,z3,'#000')
grafico.set_xlabel('eje X')
grafico.set_ylabel('eje y')
grafico.set_zlabel('eje z')
plt.title('Grafico 3D')
plt.show()
```

```python
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x4 = [100,50,80,40]
y4 = [10,50,20,40]
x4,y4 = np.meshgrid(x4,y4)

z4 = np.array([
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
])
#NOTA z4 esta sujeto si x4 y y4 mantienen una cantidad de valores alrededor de n elementos
#Si x4.size() == 4 entonces z4 sera una matriz de 4x4 4 elementos x 4 objetos internos.
figura = plt.figure()
grafico = figura.add_subplot(111,projection='3d')
auxiliar = grafico.plot_surface(x4,y4,z4,cmap=cm.viridis)
figura.colorbar(auxiliar)
plt.title('Grafico 3D')
plt.show()
```
